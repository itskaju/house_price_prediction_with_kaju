```python
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.losses import MeanSquaredError

classifier = Sequential()

# Define input layer properly
classifier.add(Input(shape=(174,)))

# First Hidden Layer
classifier.add(Dense(50, kernel_initializer='he_uniform', activation='relu'))

# Second Hidden Layer
classifier.add(Dense(25, kernel_initializer='he_uniform', activation='relu'))

# Third Hidden Layer
classifier.add(Dense(50, kernel_initializer='he_uniform', activation='relu'))

# Output Layer
classifier.add(Dense(1, kernel_initializer='he_uniform'))

# Compile the model
classifier.compile(loss=MeanSquaredError(), optimizer='adamax')
```
